```python
import os
import torch
import numpy as np
from PIL import Image
from torchvision import transforms

from model import Trainer, STNet
```

```python
trainer = Trainer(n_classes=7, lr=3e-4, backbone='resnet')
```

```python
ckpt = torch.load('./output/checkpoints_f10.343.ckpt')
trainer.model.load_state_dict(ckpt['model_state_dict'])
```

```python
patches_path = '../patches/151673'
path_list = [os.path.join(patches_path, f'{i}.jpg') for i, _ in enumerate(os.listdir(patches_path))]
```

```python
test_transforms = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
])
```

```python
features = []
for path in path_list:
    print(path)
    img = Image.open(path)
    img = np.array(img) / 255.
    img = test_transforms(img)
    feat = trainer.get_features(img.unsqueeze(dim=0).float().cuda())
    features.append(feat.detach().cpu().numpy())
features = np.array(features)
```

```python
np.save('output/151673_resnet101_class_feat_2048.npy', features)
```
